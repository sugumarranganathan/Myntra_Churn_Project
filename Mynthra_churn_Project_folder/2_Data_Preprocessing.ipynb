```python
df = pd.read_csv("fashion_customer_churn.csv")
```

```python
if "CustomerID" in df.columns:
    df = df.drop("CustomerID", axis=1)

print("First 5 rows after dropping CustomerID:")
print(df.head())
```

```python
label_encoders = {}
categorical_cols = df.select_dtypes(include=["object"]).columns.tolist()

if "Churn" in categorical_cols:
    categorical_cols.remove("Churn")

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

```python
target_encoder = LabelEncoder()
df["Churn"] = target_encoder.fit_transform(df["Churn"])

pickle.dump(label_encoders, open("fashion_rf_label_encoders.pkl", "wb"))
pickle.dump(target_encoder, open("fashion_rf_target_encoder.pkl", "wb"))

df.to_csv("preprocessed_fashion_customer_churn.csv", index=False)
```

```python
print("\nPreprocessing completed successfully.")
print("Saved files:")
print("- preprocessed_fashion_customer_churn.csv")
print("- fashion_rf_label_encoders.pkl")
print("- fashion_rf_target_encoder.pkl")
```
