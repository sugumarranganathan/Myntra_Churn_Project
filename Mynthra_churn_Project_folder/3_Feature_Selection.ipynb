```python
# Load preprocessed dataset
df = pd.read_csv("preprocessed_fashion_customer_churn.csv")

# Feature / target split
X = df.drop("Churn", axis=1)
y = df["Churn"]
```

```python
print("FEATURE SELECTION")
print("=" * 50)
print("\nSelected Features:")
print(X.columns.tolist())

print("\nTotal Number of Features:", X.shape[1])
print("\nTarget Variable: Churn")
```

```python
# Optional correlation analysis
correlation_with_target = df.corr(numeric_only=True)["Churn"].drop("Churn").sort_values(ascending=False)

print("\nFeature Correlation with Churn:")
print(correlation_with_target)
```

```python
# Save selected features list
with open("selected_features.txt", "w") as f:
    f.write("Selected Features for Churn Prediction:\n")
    for feature in X.columns:
        f.write(feature + "\n")

print("\nselected_features.txt saved successfully.")
```
